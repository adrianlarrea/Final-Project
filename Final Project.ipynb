### Defining the models

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

def evaluate_regression_models(X, y):
    models = [
        KNeighborsRegressor(),
        DecisionTreeRegressor(),
        RandomForestRegressor(),
        SVR(),
        GradientBoostingRegressor(),
        MLPRegressor()
    ]
    
    model_names = [
        "KNN",
        "Decision Tree",
        "Random Forest",
        "SVR",
        "Gradient Boosting",
        "Neural Networks"
    ]
    
    results = {}
    
    for model, name in zip(models, model_names):
        model.fit(X, y)
        y_pred = model.predict(X)
        r2 = r2_score(y, y_pred)
        results[name] = r2
        
    return results

results = evaluate_regression_models(X, y)
print(results)
```

### RUNNING THE PEARSON CORRELATION COEFFICIENT TEST

Testing if more wealth leads to more life expectancy

 H0: there is no relationship between salary and life expectancy
 H1: there is a relationship between salary and life expectancy
  
Significance level is 0.05

```python
import pandas as pd
import scipy.stats as stats


stat, p = stats.pearsonr(data['Renda mitja barri'], data['Mean_life_exp'])

print('Correlation coefficient:', stat)
print('P-value:', p)
```

The output suggests that there is a positive and weak linear relationship between the two variables in question: salary and life expectancy. The correlation coefficient of 0.49 indicates that there is a moderate positive correlation between the two variables. This means that, as one variable increases, the other variable is also likely to increase, although the relationship is not extremely strong.

The P-value of 1.28e-05 is much less than the commonly used significance level of 0.05, which means that there is a very low probability of observing this relationship by chance if the null hypothesis (that there is no relationship between salary and life expectancy) is true. This suggests that there is a statistically significant relationship between the two variables.
The since the p is low, H0 must go. We reject Null Hypothesis
